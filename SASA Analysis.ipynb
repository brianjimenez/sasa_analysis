# SASA Analysis

## 1A2K

```python
%matplotlib inline
import csv
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

with open('1A2K_freesasa.csv') as freesasa_input:
    with open('1A2K_access.csv') as access_input:
        freesasa = csv.reader(freesasa_input)
        access = csv.reader(access_input)
        
        complex_time_freesasa = [float(line[-1]) for line in freesasa]
        complex_time_access = [float(line[-1]) for line in access]
        
        # Computation time
        x = np.arange(1, 101, 1)
        plt.title('Top 100 SASA Computation time')
        plt.xlabel('Complex')
        plt.ylabel('Time (s)')
        plt.plot(x, complex_time_freesasa, label='freesasa')
        plt.plot(x, complex_time_access, label='access')
        plt.legend()
        plt.show()
        
        # Freesasa - Access correlation
        freesasa_input.seek(0)
        access_input.seek(0)
        complex_sasa_freesasa = [float(line[-2]) for line in freesasa]
        complex_sasa_access = [float(line[-2]) for line in access]
        plt.scatter(complex_sasa_freesasa, complex_sasa_access)
        plt.xlabel('freesasa')
        plt.ylabel('access')
        plt.title('SASA ($\AA^2$)')
        plt.show()
```
